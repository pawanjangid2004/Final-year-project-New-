```python
from keras.models import Sequential
from keras.layers import Conv2D, Dropout
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
class Model:
    def createAndSaveModel(self):
        classifier = Sequential()

        # add convolution
        classifier.add(Conv2D(32, (3, 3), input_shape=(64, 64, 3), activation='relu'))

        # add max pooling
        classifier.add(MaxPooling2D(pool_size=(2, 2)))

        classifier.add(Conv2D(32, (3, 3), input_shape=(64, 64, 3), activation='relu'))

        classifier.add(Dropout(1))

        # add flattening
        classifier.add(Flatten())

        # full connection

        # hidden layer
        classifier.add(Dense(units=128, activation='relu'))

        # output layer
        classifier.add(Dense(units=4, activation='softmax'))

        # compile the layers
        # create the model
        classifier.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
x_train_generator = ImageDataGenerator(rescale=1. / 255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
x_test_generator = ImageDataGenerator(rescale=1. / 255)
```

```python
x_train = x_train_generator.flow_from_directory('D:/Skin_disease_prediction/Skin_disease_prediction/train',target_size=(64, 64), batch_size=32, class_mode='categorical',
                                                        color_mode="rgb")
x_test = x_test_generator.flow_from_directory('D:/Skin_disease_prediction/Skin_disease_prediction/test',target_size=(64, 64), batch_size=32, class_mode='categorical')
```

```python
classifier = Sequential()
classifier.add(Conv2D(32, (3, 3), input_shape=(64, 64, 3), activation='relu'))

# add max pooling
classifier.add(MaxPooling2D(pool_size=(2, 2)))

classifier.add(Conv2D(32, (3, 3), input_shape=(64, 64, 3), activation='relu'))

classifier.add(Dropout(0.5))

# add flattening
classifier.add(Flatten())

# full connection

# hidden layer
classifier.add(Dense(units=128, activation='relu'))

# output layer
classifier.add(Dense(units=4, activation='softmax'))

# compile the layers
        # create the model
classifier.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
classifier.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.fit_generator(x_train, steps_per_epoch=50, epochs=100, validation_data=x_test, validation_steps=10)
```

```python
import keras
import operator
#from keras.preprocessing import image
import keras.utils as image
from keras.models import model_from_json
import numpy as np
# read the json model
file = open('my_model.json', 'r')
data = file.read()
#print(data)

file.close()

# classifier will load the model from the data
# data -> contents of the my_model.json file
classifier = model_from_json(data)

        # load waits
classifier.load_weights('weights.h5')

# load the test image
#from keras.preprocessing import image
testImageFile = "D:/Skin_disease_prediction/Skin_disease_prediction/test/Actinic keratosis/ISIC_0067078.jpg"
test_image = image.load_img(testImageFile, target_size=(64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0)

classifier.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

result = classifier.predict(test_image)
print(result)
prediction ={
            'Acne': result[0][0],
            'Actinic keratosis': result[0][1],
            'Basal cell carcinoma': result[0][2],
            'Melanoma': result[0][3]
                         }
prediction = sorted(prediction.items(), key= operator.itemgetter(1), reverse= True)
print(prediction[0])
```
